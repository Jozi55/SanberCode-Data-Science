```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
# code here
fig, ax = plt.subplots(figsize=(16,8),nrows=1, ncols=3, sharex=True, sharey=True)

ax[0].plot(x, tan, color='#0335fc', label='tanX')
ax[1].plot(x, sin, color='#fc1c03', label='sinX')
ax[2].plot(x, cos, color='green', label='cosX')
ax[0].set_title('Grafix TanX')
ax[1].set_title('Grafix sinx')
ax[2].set_title('Grafix cosx')
fig.legend(loc='upper center')
plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
# code here
fig, ax = plt.subplots(figsize=(14,8))
ax.plot(x , y, color='blue')

p1 = ax.inset_axes([-0.1,0.9,0.3,0.3])
p1.plot(x,y4, color='#d4f542')
p1.plot(x,y, color='blue')
p1.set_title('Grafkc CostX^5')

p2 = ax.inset_axes([0.9,0.9,0.3,0.3])
p2.plot(x,y4, color='black')
p2.plot(x,y, color='blue')
p2.set_title('Grafkc CostX^4')

p3 = ax.inset_axes([0.9,-0.1,0.3,0.3])
p3.plot(x,y4, color='green')
p3.plot(x,y, color='blue')
p3.set_title('Grafkc CostX^3')

p4 = ax.inset_axes([-0.1,-0.1,0.3,0.3])
p4.plot(x,y4, color='red')
p4.plot(x,y, color='blue')
p4.set_title('Grafkc CostX^2')

plt.show()
```
